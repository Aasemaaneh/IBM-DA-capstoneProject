## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# Query to retrieve 'ConvertedComp' data
QUERY = """
SELECT ConvertedComp
FROM master
"""

# Read the data into a DataFrame
df = pd.read_sql_query(QUERY, conn)

# Plot histogram for 'ConvertedComp'
plt.figure(figsize=(8, 6))
plt.hist(df['ConvertedComp'].dropna(), bins=30, color='skyblue', edgecolor='black')  # Drop NA values before plotting
plt.xlabel('Converted Compensation')
plt.ylabel('Frequency')
plt.title('Histogram of Converted Compensation')
plt.grid(True)
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# Query to retrieve 'Age' data
QUERY = """
SELECT Age
FROM master
"""

# Read the data into a DataFrame
df = pd.read_sql_query(QUERY, conn)

# Plot box plot for 'Age'
plt.figure(figsize=(8, 6))
plt.boxplot(df['Age'].dropna())  # Drop NA values before plotting
plt.xlabel('Age')
plt.title('Box Plot of Age')
plt.show()

```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# Query to retrieve 'Age' and 'WorkWeekHrs' data
QUERY = """
SELECT Age, WorkWeekHrs
FROM master
"""

# Read the data into a DataFrame
df = pd.read_sql_query(QUERY, conn)

# Plot scatter plot for 'Age' and 'WorkWeekHrs'
plt.figure(figsize=(8, 6))
plt.scatter(df['Age'], df['WorkWeekHrs'], alpha=0.5, color='blue')
plt.xlabel('Age')
plt.ylabel('Work Week Hours')
plt.title('Scatter Plot of Age vs Work Week Hours')
plt.grid(True)
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# Query to retrieve 'WorkWeekHrs', 'CodeRevHrs', and 'Age' data
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age
FROM master
"""

# Read the data into a DataFrame
df = pd.read_sql_query(QUERY, conn)

# Define bubble sizes based on the 'Age' column (scaling for better visualization)
bubble_sizes = df['Age']   # You can adjust the scaling factor as needed

# Plot bubble plot for 'WorkWeekHrs', 'CodeRevHrs', and 'Age'
plt.figure(figsize=(10, 8))
plt.scatter(df['WorkWeekHrs'], df['CodeRevHrs'], s=bubble_sizes, alpha=0.5, c='blue')
plt.xlabel('Work Week Hours')
plt.ylabel('Code Review Hours')
plt.title('Bubble Plot of Work Week Hours vs Code Review Hours (Bubble size: Age)')
plt.grid(True)
plt.show()

```

Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
QUERY = """
SELECT *
FROM DatabaseDesireNextYear
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df
df_count=df['DatabaseDesireNextYear'].value_counts().rename_axis('DatabaseName').reset_index(name='Counts')
df_count
df_top5=df_count[:5]
df_top5
plt.pie(df_top5['Counts'],labels=df_top5['DatabaseName'], autopct='%.2f')
plt.title('Top 5 databases to learn')
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
Query= """
SELECT WorkWeekHrs, CodeRevHrs,Age
FROM master
WHERE Age BETWEEN 30 AND 35 """
df=pd.read_sql_query(Query,conn)
df
# Group the data by Age and calculate the median for WorkWeekHrs and CodeRevHrs
median_data = df.groupby('Age')[['WorkWeekHrs', 'CodeRevHrs']].median()
median_data.plot(kind='bar',stacked=True)
plt.ylabel('Median Hours')
```

```python
Query= """
SELECT WorkWeekHrs, CodeRevHrs,Age
FROM master
WHERE Age BETWEEN 25 AND 30 """
df=pd.read_sql_query(Query,conn)
df
# Group the data by Age and calculate the median for WorkWeekHrs and CodeRevHrs
median_data = df.groupby('Age')[['WorkWeekHrs', 'CodeRevHrs']].median()
median_data.plot(kind='bar',stacked=True)
plt.ylabel('Median Hours')
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
QUERY = """
SELECT Age, ConvertedComp
FROM master
WHERE Age BETWEEN 45 AND 60
"""
df=pd.read_sql_query(QUERY,conn)
df_median=df.groupby('Age').median()
plt.plot(df_median)
plt.ylabel('COMP USD')
plt.xlabel('Age')
plt.title('the median `ConvertedComp` for all ages from 45 to 60.')
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
QUERY="""
SELECT MainBranch
from master"""
df=pd.read_sql_query(QUERY,conn)
dfh=df.value_counts().rename_axis('MainBranch').reset_index(name='Counts')
print(dfh)
dfh.plot(kind='barh')
plt.xlabel('Count')
plt.ylabel('MainBranch')
plt.title('Count of MainBranch')
```
